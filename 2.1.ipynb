```python
import pandas as pd
import numpy as np
housing = pd.read_csv("SD1970_housing.csv")
import matplotlib.pyplot as plot
```

```python
housing = housing[["Block group","Place Name","Total housing units"]]
housing
```

```python
housing["Total housing units"] = housing["Total housing units"].str.replace(r'[^\w\s]+', '')
housing["Total housing units"] = housing["Total housing units"].astype(int)
housing = housing[housing["Total housing units"] > 0]
```

```python
housing
```

```python
housing["Place Name"].value_counts()
```

```python
housing = housing.groupby(["Place Name"]).sum()["Total housing units"].sort_values(ascending = False)
#How big is each residential area?
#Second common block: Chula Vista
housing
```
