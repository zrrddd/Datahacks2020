```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
import matplotlib.pyplot as plot
```

```python
from sklearn.model_selection import train_test_split 
from sklearn import metrics
%matplotlib inline
```

```python
population = pd.read_csv("SD1970_population.csv")
```

```python
def extract(x):
    x = x.split(" ")[2]
    return math.floor(float(x))
```

```python
Gender_ratio = population[population.columns[0:48]]

Gender_ratio["Census Tract Name"] = Gender_ratio["Census Tract Name"].apply(extract)

Gender_ratio = Gender_ratio.drop(["Place Name"], axis = 1)

Gender_ratio = Gender_ratio[Gender_ratio["Male persons age under 5 years"]!="..."]

for i in list(Gender_ratio.columns[2:]):
    Gender_ratio[i] = Gender_ratio[i].str.replace(r'[^\w\s]+', '')
    Gender_ratio[i] = Gender_ratio[i].astype(int)

Gender_ratio["Total men"] = Gender_ratio.iloc[:,3:25].sum(axis = 1)

Gender_ratio = Gender_ratio[["Census Tract Name", "Block Group", "Total persons", "Total men"]]
```

```python
Gender_ratio = Gender_ratio.groupby(["Census Tract Name", "Block Group"]).sum()
```

```python
Gender_ratio["Men Ratio"] = Gender_ratio["Total men"]/Gender_ratio["Total persons"]
```

```python
Gender_ratio
```

```python
#Gender_ratio["Men Ratio"] = Gender_ratio["Total men"]/Gender_ratio["Total persons"]

Gender_ratio = Gender_ratio[Gender_ratio["Total persons"]!=0]

Gender_ratio = Gender_ratio.drop(["Total persons", "Total men"], axis = 1)
```

```python
Gender_ratio
```

```python
temp1 = population[["Census Tract Name", "Block Group", "Total persons"]]

temp2 = population[population.columns[85:95]]

Marriage_ratio = temp1.join(temp2)

Marriage_ratio["Census Tract Name"] = Marriage_ratio["Census Tract Name"].apply(extract)

Marriage_ratio = Marriage_ratio[Marriage_ratio["Total persons"]!="0"]
Marriage_ratio = Marriage_ratio.replace("...", "0")

for i in list(Marriage_ratio.columns[2:]):
    Marriage_ratio[i] = Marriage_ratio[i].str.replace(r'[^\w\s]+', '')

Marriage_ratio["Male persons now married (excludes separated)"] = Marriage_ratio["Male persons now married (excludes separated)"].astype(int)

for i in list(Marriage_ratio.columns[2:]):
    Marriage_ratio[i] = Marriage_ratio[i].astype(int)

Marriage_ratio["Married Men"] = Marriage_ratio.iloc[:, 3:4].sum(axis = 1)

Marriage_ratio["Other Men"] = Marriage_ratio.iloc[:, 4:8].sum(axis = 1)

Marriage_ratio["Married Women"] = Marriage_ratio.iloc[:, 8:9].sum(axis = 1)

Marriage_ratio["Other Women"] = Marriage_ratio.iloc[:, 9:13].sum(axis = 1)

Marriage_ratio = Marriage_ratio[["Census Tract Name", "Block Group", "Married Men", "Other Men", "Married Women","Other Women"]]
```

```python
Marriage_ratio
```

```python
Marriage_ratio = Marriage_ratio.groupby(["Census Tract Name", "Block Group"]).sum()
```

```python
Marriage_ratio["Married Ratio"] = (Marriage_ratio["Married Men"] + Marriage_ratio["Married Women"])/(Marriage_ratio["Married Men"] + Marriage_ratio["Married Women"]+ Marriage_ratio["Other Men"]+Marriage_ratio["Other Women"])
```

```python
Marriage_ratio = Marriage_ratio.drop(["Married Men", "Other Men", "Other Women", "Married Women"], axis = 1)
```

```python
Marriage_ratio = Marriage_ratio[Marriage_ratio["Married Ratio"] >= 0]
```

```python
Marriage_ratio
```

```python
df = population[population.columns[0:2]]
population = population[population.columns[80:85]]
population = population.replace('...', "0")
Race_ratio = df.join(population)

Race_ratio["Census Tract Name"] = Race_ratio["Census Tract Name"].apply(extract)

for i in Race_ratio.columns[2:]:
    Race_ratio[i] = Race_ratio[i].str.replace(r'[^\w\s]+', '')
    Race_ratio[i] = Race_ratio[i].astype(int)

Race_ratio["Total person"] = Race_ratio.iloc[:, 2:7].sum(axis = 1)
```

```python
Race_ratio
```

```python
Race_ratio = Race_ratio.groupby(["Census Tract Name", "Block Group"]).sum()
```

```python
Race_ratio["White ratio"] = Race_ratio["White persons"]/Race_ratio["Total person"]
```

```python
Race_ratio.columns
```

```python
Race_ratio = Race_ratio.drop(["Black persons", "White persons", "Indian persons", "Other specified race persons", 'Reported "other race" persons', "Total person"], axis = 1)
```

```python
Race_ratio = Race_ratio[Race_ratio["White ratio"] >= 0]
```

```python
Race_ratio
```

```python
Merged = pd.merge(Gender_ratio, Marriage_ratio, on=["Census Tract Name", "Block Group"])
```

```python
Merged = pd.merge(Merged, Race_ratio, on=["Census Tract Name", "Block Group"])
```

```python
housing = pd.read_csv("SD1970_housing.csv")
housing = housing[housing["Place Name"] == "San Diego"]

housing = housing[["Tract name", "Block group","Total housing units", " Total owner occupied average $ value of housing units "]]
housing.columns = ["Census Tract Name","Block Group","Total units", "Average price"]
housing = housing[housing["Total units"]!="0"]

housing["Average price"] = housing["Average price"].str.replace(r'[^\w\s]+', '')
housing["Average price"] = housing["Average price"].str.rstrip()
housing["Filter"] = housing["Average price"].str.isdigit()
housing = housing[housing["Filter"] == True]

housing["Average price"] = housing["Average price"].str.replace(r'[^\w\s]+', '')
housing["Average price"] = housing["Average price"].str.rstrip()
housing["Average price"] = housing["Average price"].astype(int)
housing["Average price"] = housing["Average price"]/100

housing["Total units"] = housing["Total units"].str.replace(r'[^\w\s]+', '')
housing["Total units"] = housing["Total units"].astype(int)

housing["Total price"] = housing["Total units"] * housing["Average price"]
```

```python
housing_group = housing[["Census Tract Name", "Block Group", "Total units","Total price"]]
housing_group["Census Tract Name"] = housing_group["Census Tract Name"].apply(extract)
```

```python
housing_group = housing_group.groupby(["Census Tract Name", "Block Group"]).sum()
housing_group["Average price"] = housing_group["Total price"]/housing_group["Total units"]
```

```python
Merged = pd.merge(Merged, housing_group, on=["Census Tract Name", "Block Group"])
```

```python
Merged = Merged.drop(["Total units", "Total price"], axis = 1)
```

```python
Merged["Average price"] = Merged['Average price']/10000
```

```python
Merged
```

```python
reg = linear_model.LinearRegression()
reg.fit(Merged[["Men Ratio", "Married Ratio", "White ratio"]], Merged["Average price"])
```

```python
#R square
reg.score(Merged[["Men Ratio", "Married Ratio", "White ratio"]], Merged["Average price"])
```

```python
#Prediction for first 10
predictions = reg.predict(Merged[["Men Ratio", "Married Ratio", "White ratio"]])
predictions[0:10]
```

```python
Merged["Prediction"] = predictions
Merged
```

```python
Merged["residual"] = Merged["Prediction"] - Merged["Average price"]
Merged
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Merged["Average price"], Merged["Prediction"]))  
print('Mean Squared Error:', metrics.mean_squared_error(Merged["Average price"], Merged["Prediction"]))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Merged["Average price"], Merged["Prediction"])))
```

```python
Mean_Men_Ratio = (Merged['Men Ratio'].sum())/396
Merged["square Men Ratio distance to mean"] = (Merged['Men Ratio'] - Mean_Men_Ratio)**2
```

```python
Mean_White_ratio = (Merged['White ratio'].sum())/396
Merged["square White Ratio distance to mean"] = (Merged['White ratio'] - Mean_White_ratio)**2
Mean_Married_ratio = (Merged['Married Ratio'].sum())/396
Merged["square Married Ratio distance to mean"] = (Merged['Married Ratio'] - Mean_Married_ratio)**2
Merged
```

```python
variance_Men = (1/396)*Merged["square Men Ratio distance to mean"].sum()
std_Men = variance_Men**(1/2)
std_Men
```

```python
variance_White = (1/396)*Merged["square White Ratio distance to mean"].sum()
std_White = variance_White**(1/2)
std_White
```

```python
variance_Married = (1/396)*Merged["square Married Ratio distance to mean"].sum()
std_Married = variance_Married**(1/2)
reg.coef_
```

```python
t_stat_men_ratio = reg.coef_[0]/std_Men
t_stat_married_ratio = reg.coef_[1]/std_Married
t_stat_White_ratio = reg.coef_[2]/std_White
print('men_ratio_test:', t_stat_men_ratio)  
print('married_ratio_test:', t_stat_married_ratio)  
print('white_ratio_test:', t_stat_White_ratio)
```
