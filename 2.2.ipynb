```python
import pandas as pd
import numpy as np
import math
housing = pd.read_csv("SD1970_housing.csv")
housing = housing[housing["Place Name"] == "San Diego"]

housing = housing[["Tract name", "Block group","Total housing units", " Total owner occupied average $ value of housing units "]]
housing.columns = ["Tract","Block","Total units", "Average price"]
housing = housing[housing["Total units"]!="0"]

housing["Average price"] = housing["Average price"].str.replace(r'[^\w\s]+', '')
housing["Average price"] = housing["Average price"].str.rstrip()
housing["Filter"] = housing["Average price"].str.isdigit()
housing = housing[housing["Filter"] == True]

housing["Average price"] = housing["Average price"].str.replace(r'[^\w\s]+', '')
housing["Average price"] = housing["Average price"].str.rstrip()
housing["Average price"] = housing["Average price"].astype(int)
housing["Average price"] = housing["Average price"]/100

housing["Total units"] = housing["Total units"].str.replace(r'[^\w\s]+', '')
housing["Total units"] = housing["Total units"].astype(int)

housing["Total price"] = housing["Total units"] * housing["Average price"]
housing
```

```python
average = housing["Total price"].sum()/housing["Total units"].sum()
```

```python
average
```

```python
def extract(x):
    x = x.split(" ")[2]
    return math.floor(float(x))
```

```python
housing_group = housing[["Tract", "Block", "Total units","Total price"]]
housing_group["Tract"] = housing_group["Tract"].apply(extract)
housing_group
```

```python
housing_group_byblock = housing_group[["Block","Total units","Total price"]]
housing_group_byblock = housing_group_byblock.groupby(["Block"]).sum()
housing_group_byblock["Average price"] = housing_group_byblock["Total price"]/housing_group_byblock["Total units"]
```

```python
housing_group_byblock.sort_values("Average price", ascending = False)
#Block 8 has the higest average price 
```

```python
housing_group = housing_group.groupby(["Tract", "Block"]).sum()
housing_group["Average price"] = housing_group["Total price"]/housing_group["Total units"]
```
