```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
import matplotlib.pyplot as plot
```

```python
population = pd.read_csv("SD1970_population.csv")
```

```python
def extract(x):
    x = x.split(" ")[2]
    return math.floor(float(x))
```

```python
df = population[population.columns[0:2]]
population = population[population.columns[80:85]]
population = population.replace('...', "0")
Race_ratio = df.join(population)

Race_ratio["Census Tract Name"] = Race_ratio["Census Tract Name"].apply(extract)

for i in Race_ratio.columns[2:]:
    Race_ratio[i] = Race_ratio[i].str.replace(r'[^\w\s]+', '')
    Race_ratio[i] = Race_ratio[i].astype(int)

Race_ratio["Total person"] = Race_ratio.iloc[:, 2:7].sum(axis = 1)

Race_ratio["White ratio"] = Race_ratio["White persons"]/Race_ratio["Total person"]

Race_ratio = Race_ratio[["Census Tract Name", "Block Group", "White ratio"]]

Race_ratio = Race_ratio[Race_ratio["White ratio"] >= 0]

Race_ratio.sort_values("White ratio", ascending = False)
```
